```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
patents=pd.read_csv('../patetnts/q2_patents_topics.csv') 
```

```python
patents.head()
```

```python
patents['publication date'] = pd.to_datetime(patents['publication date'])
```

```python
articles = pd.read_csv('articles_scopus_q2_topics.csv')
```

```python
articles.head()
```

# Some Counting

```python
ax = sns.countplot(y="topic", data=patents).set_title("Patents topic count")
```

```python
ax = sns.countplot(y="class", data=patents).set_title("Patents class count")
```

```python
ax = sns.countplot(y="topic", hue="class", data=patents).set_title("Patents topic + class count")
```

```python
ax = sns.countplot(y="topic", data=articles).set_title("articles topic count")
```

```python
patents['Year'] = pd.DatetimeIndex(patents['publication date']).year
```

```python
articlesCntByYear = articles.groupby(['topic', 'Year']).size().reset_index(name='counts')
```

```python
patentsTopicCntByYear = patents.groupby(['topic', 'Year']).size().reset_index(name='counts')
patentsClassCntByYear = patents.groupby(['class', 'Year']).size().reset_index(name='counts')
```

```python
patentsClassCntByYear
```

```python
articlesCntByYear = articlesCntByYear[articlesCntByYear['Year'] != 2021]
patentsTopicCntByYear = patentsTopicCntByYear[patentsTopicCntByYear['Year'] != 2021]
patentsClassCntByYear = patentsClassCntByYear[patentsClassCntByYear['Year'] != 2021]
```

# TREND ANALYSIS

 # Articles Trend

```python
sns.lineplot(data=articlesCntByYear, x="Year", y="counts", hue="topic", legend="full", palette="bright").set_title("articles topic trend")
```

```python
for i in range(8):
    plt.figure(i)
    s=sns.lineplot(data=articlesCntByYear[articlesCntByYear['topic'] == i], x="Year", y="counts", hue="topic", legend="full", palette="bright").set_title("articles topic " + str(i) +" trend")
    
```

# Patent Analysis 

```python
sns.lineplot(data=patentsTopicCntByYear, x="Year", y="counts", hue="topic", legend="full", palette="bright").set_title("patents topic trend")
```

```python
for i in range(8):
    plt.figure(i)
    s=sns.lineplot(data=patentsTopicCntByYear[patentsTopicCntByYear['topic'] == i], x="Year", y="counts", hue="topic", legend="full", palette="bright").set_title("patents topic " + str(i) +" trend")
```

```python
sns.lineplot(data=patentsClassCntByYear, x="Year", y="counts", hue="class", legend="full", palette="bright").set_title("patents class trend")
```

```python
allDocumentsByYear = pd.concat([articlesCntByYear, patentsTopicCntByYear]).groupby(['topic', 'Year'],as_index=False).sum()
allDocumentsByYear
```

# All Documents Analysis

```python
sns.lineplot(data=allDocumentsByYear, x="Year", y="counts", hue="topic", legend="full", palette="bright").set_title("all documents topic trend")
```

```python
for i in range(8):
    plt.figure(i)
    s=sns.lineplot(data=allDocumentsByYear[allDocumentsByYear['topic'] == i], x="Year", y="counts", hue="topic", legend="full", palette="bright").set_title("all Documents topic " + str(i) +" trend")
```
